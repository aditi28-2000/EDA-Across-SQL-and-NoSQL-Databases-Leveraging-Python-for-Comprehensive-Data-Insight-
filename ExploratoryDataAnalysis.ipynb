```python
# Rename the value columns to heavy_veh and light_veh respectively to apply the agg_dict
traffic_heavy_df.rename(columns={"value": "heavy_veh"}, inplace=True)
traffic_light_df.rename(columns={"value": "light_veh"}, inplace=True)

# Setting 'timestamp' as the index for all datasets
traffic_heavy_df.set_index('timestamp', inplace=True)
traffic_light_df.set_index('timestamp', inplace=True)
environment_df.set_index('timestamp', inplace=True)

# Combining environmental data with the traffic datasets
environ_traffic=pd.concat([traffic_heavy_df, traffic_light_df, environment_df],axis=1)
```

```python
# define aggregation-dict
agg_dict = {"temperature": "max", "humidity": "max", "sunshine": "sum", 
            "light_veh": "sum", "heavy_veh": "sum",
            }

# resampling the combined environ_traffic dataset to "1h" and applying the aggregation-dict to it
environ_traffic = environ_traffic.resample("1h").agg(agg_dict)
```
